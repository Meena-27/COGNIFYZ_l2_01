```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Zomato dataset
zomato_data = pd.read_csv("/content/zomato.csv")

# Assuming the dataset has columns 'Aggregate rating' and 'num of ratings'
# containing the aggregate rating and the number of ratings respectively

# Visualize the distribution of aggregate ratings using a histogram
plt.figure(figsize=(8, 6))
plt.hist(zomato_data['rate (out of 5)'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Aggregate Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Determine the most common rating range
most_common_rating_range = zomato_data['rate (out of 5)'].mode()[0]
print("Most common rating range:", most_common_rating_range)

# Calculate the average number of votes received by restaurants
average_num_of_votes = zomato_data['num of ratings'].mean()
print("Average number of votes received by restaurants:", average_num_of_votes)
```
